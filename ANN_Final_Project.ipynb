## Discriminator

The Discriminator of the Style-GAN is implemented in the same way as in the progressive GAN. 

```
class Discriminator(Model):

  def __init__(self):

    super(Discriminator, self).__init__()

    #input_shape = (1, 1024, 1024, 3)
    
    # 1x1 convolution layer !!! BUT CHECK AGAIN WITH PAPER, IT IS 4x4 FOR PROGRESSIVE GAN BUT 8x8 FOR STYLE GAN !!! Or is that just input size?!
    self.conv_1 = layers.Conv2D(filters = 16,
                                kernel_size = 1,
                                input_shape = [512,512,3])
    # Leaky ReLU activation with leakiness 0.2                           
    self.LReLU_1 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_2 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_3 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_4 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_5 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_6 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_7 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_8 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_9 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_10 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_11 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_12 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_13 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_14 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_15 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_16 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_17 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_18 = layers.LeakyReLU(alpha = 0.2)
    self.LReLU_19 = layers.LeakyReLU(alpha = 0.2)

    # 3x3 convolution layers
    self.conv_16 = layers.Conv2D(filters = 16,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_32 = layers.Conv2D(filters = 32,
                                kernel_size= 3,
                                padding = 'same')
    self.conv_32_2 = layers.Conv2D(filters = 32,
                                   kernel_size = 3,
                                   padding = 'same')
    # Downsample
    self.downsample = layers.AveragePooling2D(pool_size=(2,2),
                                                 strides = (2,2))
    self.downsample_2 = layers.AveragePooling2D(pool_size=(2,2),
                                                 strides = (2,2))
    self.downsample_3 = layers.AveragePooling2D(pool_size=(2,2),
                                                 strides = (2,2))
    self.downsample_4 = layers.AveragePooling2D(pool_size=(2,2),
                                                 strides = (2,2))
    self.downsample_5 = layers.AveragePooling2D(pool_size=(2,2),
                                                 strides = (2,2))
    self.downsample_6 = layers.AveragePooling2D(pool_size=(2,2),
                                                 strides = (2,2))
    self.downsample_7 = layers.AveragePooling2D(pool_size=(2,2),
                                                 strides = (2,2))
    self.downsample_8 = layers.AveragePooling2D(pool_size=(2,2),
                                                 strides = (2,2))
    
    self.conv_64 = layers.Conv2D(filters = 64,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_64_2 = layers.Conv2D(filters = 64,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_128 = layers.Conv2D(filters = 128,
                                  kernel_size = 3,
                                  padding = 'same')
    self.conv_128_2 = layers.Conv2D(filters = 128,
                                  kernel_size = 3,
                                  padding = 'same')
    self.conv_256 = layers.Conv2D(filters = 256,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_256_2 = layers.Conv2D(filters = 256,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_512 = layers.Conv2D(filters = 512,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_512_2 = layers.Conv2D(filters = 512,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_512_3 = layers.Conv2D(filters = 512,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_512_4 = layers.Conv2D(filters = 512,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_512_5 = layers.Conv2D(filters = 512,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_512_6 = layers.Conv2D(filters = 512,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_512_7 = layers.Conv2D(filters = 512,
                                kernel_size = 3,
                                padding = 'same')
    self.conv_512_8 = layers.Conv2D(filters = 512,
                                kernel_size = 3,
                                padding = 'same')
    # ??? self.minibatch_std = layers.
    self.conv_last = layers.Conv2D(filters = 512,
                                   kernel_size = 4)
    self.dense = layers.Dense(units = 1)

  @tf.function
  def call(self, x):
    x = self.conv_1(x)
    x = self.LReLU_1(x)
    x = self.conv_16(x)
    x = self.LReLU_2(x)
    x = self.conv_32(x)
    x = self.LReLU_3(x)
    x = self.downsample(x)

    x = self.conv_32_2(x)
    x = self.LReLU_4(x)
    x = self.conv_64(x)
    x = self.LReLU_5(x)
    x = self.downsample_2(x)

    x = self.conv_64_2(x)
    x = self.LReLU_6(x)
    x = self.conv_128(x)
    x = self.LReLU_7(x)
    x = self.downsample_3(x)

    x = self.conv_128_2(x)
    x = self.LReLU_8(x)
    x = self.conv_256(x)
    x = self.LReLU_9(x)
    x = self.downsample_4(x)

    x = self.conv_256_2(x)
    x = self.LReLU_10(x)
    x = self.conv_512(x)
    x = self.LReLU_11(x)
    x = self.downsample_5(x)

    x = self.conv_512_2(x)
    x = self.LReLU_12(x)
    x = self.conv_512_3(x)
    x = self.LReLU_13(x)
    x = self.downsample_6(x)

    x = self.conv_512_4(x)
    x = self.LReLU_14(x)
    x = self.conv_512_5(x)
    x = self.LReLU_15(x)
    x = self.downsample_7(x)

    x = self.conv_512_6(x)
    x = self.LReLU_16(x)
    x = self.conv_512_7(x)
    x = self.LReLU_17(x)
    x = self.downsample_8(x)

   # x = self.minibatch_std(x)
    x = self.conv_512_8(x)
    x = self.LReLU_18(x)
    x = self.conv_last(x)
    x = self.LReLU_19(x)
    x = self.dense(x)

    return x
```

```
inp = tf.zeros((1, 1024, 1024, 3))

disc = Discriminator()
result = disc(inp)

print(result.shape)
disc.summary()
```

# Mapping Network *f*

* 8-layer MLP, fully connected, with 512 units each (including z and w)
* activation is Leaky ReLU with $\alpha = 0.2$

* to make the training more stable "we reduce the learning rate by two orders of magnitude for the mapping network, i.e., $\lambda ' = 0.01 * \lambda$
* initialization of weights: $\mathcal{N}(0,1)$


```
class Mapping_network(Model):

  def __init__(self):
    super(Mapping_network, self).__init__()

    self.layerBlock = [layers.Dense(units = 512),
                       layers.LeakyReLU(0.2),
                       layers.Dense(units = 512),
                       layers.LeakyReLU(0.2),
                       layers.Dense(units = 512),
                       layers.LeakyReLU(0.2),
                       layers.Dense(units = 512),
                       layers.LeakyReLU(0.2),
                       layers.Dense(units = 512),
                       layers.LeakyReLU(0.2),
                       layers.Dense(units = 512),
                       layers.LeakyReLU(0.2),
                       layers.Dense(units = 512),
                       layers.LeakyReLU(0.2),
                       layers.Dense(units = 512),
                       layers.LeakyReLU(0.2),
    ]

  def call(self, x):
    for layer in self.layerBlock:
      x = layer(x)
    return x
```

```
model = Mapping_network()
t = model(tf.zeros([1, 28, 28, 1]))
model.summary()
```
